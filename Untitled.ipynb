```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# 🔄 Convert Date/Time columns to datetime format
df_petroleum['Date'] = pd.to_datetime(df_petroleum['Date'])
df_gas['Time'] = pd.to_datetime(df_gas['Time'])

# 🏗️ Feature Engineering (Add Year & Month)
df_petroleum['Year'] = df_petroleum['Date'].dt.year
df_petroleum['Month'] = df_petroleum['Date'].dt.month
df_gas['Year'] = df_gas['Time'].dt.year
df_gas['Month'] = df_gas['Time'].dt.month

# 🎯 Define Features (X) and Target (y)
X_petroleum = df_petroleum.drop(columns=['Date', 'Total flow'])  # Features
y_petroleum = df_petroleum['Total flow']  # Target

X_gas = df_gas.drop(columns=['Time', 'Production'])  # Features
y_gas = df_gas['Production']  # Target

# 🧪 Train-test split
X_train_p, X_test_p, y_train_p, y_test_p = train_test_split(X_petroleum, y_petroleum, test_size=0.2, random_state=42)
X_train_g, X_test_g, y_train_g, y_test_g = train_test_split(X_gas, y_gas, test_size=0.2, random_state=42)

# 🔄 Normalize the data
scaler_p = MinMaxScaler()
X_train_p_scaled = scaler_p.fit_transform(X_train_p)
X_test_p_scaled = scaler_p.transform(X_test_p)

scaler_g = MinMaxScaler()
X_train_g_scaled = scaler_g.fit_transform(X_train_g)
X_test_g_scaled = scaler_g.transform(X_test_g)

print("✅ Data Preprocessing Done!")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# 📌 Function to Train and Evaluate Models
def train_and_evaluate(model, X_train, X_test, y_train, y_test, model_name):
    model.fit(X_train, y_train)  # Train Model
    y_pred = model.predict(X_test)  # Make Predictions
    
    # 📊 Evaluation Metrics
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    
    print(f"📌 {model_name} Performance:")
    print(f"🔹 MAE: {mae:.4f}")
    print(f"🔹 RMSE: {rmse:.4f}")
    print(f"🔹 R² Score: {r2:.4f}")
    print("-" * 40)
    
    return y_pred

# ✅ Train & Evaluate Models for Petroleum Flow Prediction
print("\n🚀 Training Petroleum Flow Models...\n")
linreg_pred_p = train_and_evaluate(LinearRegression(), X_train_p_scaled, X_test_p_scaled, y_train_p, y_test_p, "Linear Regression")
rf_pred_p = train_and_evaluate(RandomForestRegressor(n_estimators=100, random_state=42), X_train_p_scaled, X_test_p_scaled, y_train_p, y_test_p, "Random Forest")
xgb_pred_p = train_and_evaluate(XGBRegressor(n_estimators=100, random_state=42), X_train_p_scaled, X_test_p_scaled, y_train_p, y_test_p, "XGBoost")

# ✅ Train & Evaluate Models for Gas Production Prediction
print("\n🚀 Training Gas Production Models...\n")
linreg_pred_g = train_and_evaluate(LinearRegression(), X_train_g_scaled, X_test_g_scaled, y_train_g, y_test_g, "Linear Regression")
rf_pred_g = train_and_evaluate(RandomForestRegressor(n_estimators=100, random_state=42), X_train_g_scaled, X_test_g_scaled, y_train_g, y_test_g, "Random Forest")
xgb_pred_g = train_and_evaluate(XGBRegressor(n_estimators=100, random_state=42), X_train_g_scaled, X_test_g_scaled, y_train_g, y_test_g, "XGBoost")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Function to Plot Actual vs Predicted Values
def plot_actual_vs_pred(y_test, y_pred, title):
    plt.figure(figsize=(8, 5))
    plt.scatter(y_test, y_pred, color="blue", alpha=0.5)
    plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color="red", linestyle="dashed")
    plt.xlabel("Actual Values")
    plt.ylabel("Predicted Values")
    plt.title(title)
    plt.show()

# 📌 Plot Actual vs Predicted for Petroleum Flow
plot_actual_vs_pred(y_test_p, rf_pred_p, "Actual vs Predicted (Random Forest - Petroleum Flow)")
plot_actual_vs_pred(y_test_p, xgb_pred_p, "Actual vs Predicted (XGBoost - Petroleum Flow)")

# 📌 Plot Actual vs Predicted for Gas Production
plot_actual_vs_pred(y_test_g, rf_pred_g, "Actual vs Predicted (Random Forest - Gas Production)")
plot_actual_vs_pred(y_test_g, xgb_pred_g, "Actual vs Predicted (XGBoost - Gas Production)")

# 📌 Residual Plot (Errors)
def plot_residuals(y_test, y_pred, title):
    residuals = y_test - y_pred
    plt.figure(figsize=(8, 5))
    sns.histplot(residuals, bins=30, kde=True, color="purple")
    plt.xlabel("Residuals (Errors)")
    plt.title(title)
    plt.show()

# 🔍 Residuals for Best Model (Petroleum)
plot_residuals(y_test_p, rf_pred_p, "Residual Plot (Random Forest - Petroleum Flow)")

# 🔍 Residuals for Best Model (Gas)
plot_residuals(y_test_g, rf_pred_g, "Residual Plot (Random Forest - Gas Production)")

# 📌 Feature Importance (Random Forest)
rf_model_p = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model_p.fit(X_train_p_scaled, y_train_p)
feature_importance = rf_model_p.feature_importances_

plt.figure(figsize=(8, 5))
sns.barplot(x=feature_importance, y=X_train_p.columns, palette="viridis")
plt.xlabel("Feature Importance Score")
plt.ylabel("Features")
plt.title("Feature Importance (Random Forest - Petroleum Flow)")
plt.show()
```
